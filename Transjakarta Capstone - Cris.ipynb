# Data
Untuk menjawab pertanyaan di atas, kita akan menganalisa data Transjakarta

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Data Transjakarta memiliki struktur sebagai berikut:
* transID: Unique transaction id for every transaction
* payCardID: Customers main identifier. The card customers use as a ticket for entrance and exit.
* payCardBank: Customers card bank issuer name
* payCardName: Customers name that is embedded in the card.
* payCardSex: Customers sex that is embedded in the card
* payCardBirthDate: Customers birth year
* corridorID: Corridor ID / Route ID as key for route grouping.
* corridorName: Corridor Name / Route Name contains Start and Finish for each route.
* direction: 0 for Go, 1 for Back. Direction of the route.
* tapInStops: Tap In (entrance) Stops ID for identifying stops name
* tapInStopsName: Tap In (entrance) Stops Name where customers tap in.
* tapInStopsLat: Latitude of Tap In Stops
* tapInStopsLon: Longitude of Tap In Stops
* stopStartSeq: Sequence of the stops, 1st stop, 2nd stops etc. Related to direction.
* tapInTime: Time of tap in. Date and time
* tapOutStops: Tap Out (Exit) Stops ID for identifying stops name
* tapOutStopsName: Tap out (exit) Stops Name where customers tap out.
* tapOutStopsLat: Latitude of Tap Out Stops
* tapOutStopsLon: Longitude of Tap Out Stops
* stopEndSeq: Sequence of the stops, 1st stop, 2nd stops etc. Related to direction.
* tapOutTime: Time of tap out. Date and time
* payAmount: The number of what customers pay. Some are free. Some not.

# Missing Value

```python
# persentase missing value di tiap kolom
df.isna().sum()/df.shape[0]*100
```

```python
# Sebaran missing value di dalam data
plt.figure(figsize=(10,10))
sns.heatmap(df[['corridorID','corridorName','tapOutStops','tapOutStopsName','tapOutTime','tapOutStopsLat','tapOutStopsLon','stopEndSeq','tapInStops','payAmount']].isna())
```

Secara garis besar:  
* *missing value* di yang terdapat pada data tidaklah begitu besar. Kolom `tapOutStops` memiliki *missing value* terbesar dengan persentase `6.03%`
* *missing value* di kolom `tapOutStops`, `tapOutStopsName`, `tapOutTime`, `tapOutStopsLat`, `tapOutStopsLon`, dan `stopEndSeq` memiliki sebaran yang terlihat berkaitan, jika data di `tapOutStops` kosong, maka kemungkinan besar data di kolom `tapOutStopsName`, `tapOutTime`, `tapOutStopsLat`, `tapOutStopsLon`, dan `stopEndSeq` juga kosong.  
* *missing value* yang berkaitan tidak memberikan informasi tambahan. Karena seharusnya kalau tapIn akan tapOut

Cara untuk menangani *missing value*:
* pertama, menghapus baris/kolom yang berisi *missing value*. Cara ini mungkin akan dilakukan karena jumlah *missing value* kecil.

# Menghapus *missing value*

```python
# Menghapus baris dengan missing values
data_cleaned = df.dropna()

# Menampilkan ukuran dataframe sebelum dan sesudah penghapusan
print("Ukuran data sebelum pembersihan:", df.shape)
print("Ukuran data setelah pembersihan:", data_cleaned.shape)
```

# Cek Distribusi Data

```python
# Menampilkan statistik deskriptif untuk setiap kolom
print(data_cleaned.describe(include='all'))
```

```python
# Histogram untuk kolom `payAmount`
plt.figure(figsize=(10,6))
sns.histplot(data_cleaned['payAmount'], bins=30, kde=True)
plt.title('Distribusi Jumlah Pembayaran')
plt.show()

# Boxplot untuk kolom `payAmount`
plt.figure(figsize=(10,6))
sns.boxplot(data_cleaned['payAmount'])
plt.title('Boxplot Jumlah Pembayaran')
plt.show()

# Bar chart untuk kolom `payCardBank`
plt.figure(figsize=(12,7))
sns.countplot(y=data_cleaned['payCardBank'], order=data_cleaned['payCardBank'].value_counts().index)
plt.title('Distribusi Transaksi Berdasarkan Bank')
plt.show()
```

Heatmap ini menunjukkan korelasi antara variabel numerik. Terlihat ada beberapa variabel dengan korelasi positif yang kuat, tetapi kebanyakan variabel tampak tidak terlalu berkorelasi satu sama lain. Ini menunjukkan bahwa kebanyakan variabel beroperasi secara independen satu sama lain.

# Analisis Rute Populer

```python
# Creating a new column for route combining tapInStopsName and tapOutStopsName
data_cleaned['route'] = data_cleaned['tapInStopsName'] + " - " + data_cleaned['tapOutStopsName']

# Top 10 most popular routes
top_routes = data_cleaned['route'].value_counts().head(10)

# Visualizing the top routes
plt.figure(figsize=(12,8))
sns.barplot(x=top_routes.values, y=top_routes.index, palette='viridis')
plt.title('10 Rute Paling Populer')
plt.xlabel('Jumlah Transaksi')
plt.ylabel('Rute')
plt.show()
```

# Analisis Durasi Perjalanan

```python
# Calculate travel duration
data_cleaned['tapInTime'] = pd.to_datetime(data_cleaned['tapInTime'], errors='coerce')
data_cleaned['tapOutTime'] = pd.to_datetime(data_cleaned['tapOutTime'], errors='coerce')
data_cleaned['duration'] = (data_cleaned['tapOutTime'] - data_cleaned['tapInTime']).dt.total_seconds() / 60  # in minutes

# Visualize the distribution of travel duration
plt.figure(figsize=(14,6))
sns.histplot(data_cleaned['duration'], bins=100, color='lightcoral', kde=True)
plt.title('Distribusi Durasi Perjalanan (dalam menit)')
plt.xlabel('Durasi (menit)')
plt.ylabel('Jumlah Transaksi')
plt.xlim(0, 180)  # limiting to 3 hours for better visualization
plt.show()
```

# Analisis Biaya vs Durasi

```python
# Visualize the relationship between cost and travel duration using the correct column name
plt.figure(figsize=(14,7))
sns.scatterplot(x=data_cleaned['duration'], y=data_cleaned['payAmount'], hue=data_cleaned['payCardSex'], alpha=0.6, palette='viridis')
plt.title('Hubungan Biaya dengan Durasi Perjalanan')
plt.xlabel('Durasi (menit)')
plt.ylabel('Biaya Transaksi (IDR)')
plt.xlim(0, 300)  # Limiting to 5 hours for better visualization
plt.ylim(0, 10000)  # Assuming a reasonable transaction amount
plt.legend(title='Jenis Kelamin')
plt.show()
```

# Analisis Transaksi dalam seminggu

```python
# Extract day of week from the tapInTime column
data_cleaned['day_of_week'] = data_cleaned['tapInTime'].dt.dayofweek

# Mapping day of week to actual day names
days = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
data_cleaned['day_name'] = data_cleaned['day_of_week'].map(days)

# Visualize the distribution of transactions based on day of the week
plt.figure(figsize=(12,6))
sns.countplot(data=data_cleaned, x='day_name', order=days.values(), palette='coolwarm')
plt.title('Distribusi Transaksi berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.show()
```

Ada penurunan jumlah transaksi pada akhir pekan, yaitu hari Sabtu dan Minggu. Ini menunjukkan bahwa Transjakarta lebih sering digunakan untuk keperluan sehari-hari seperti bekerja atau bersekolah dibandingkan dengan kegiatan akhir pekan.


# Analisis Berdasarkan Stasiun

```python
# Top 10 stations for tap in
top_tapin_stations = data_cleaned['tapInStopsName'].value_counts().head(10)

# Top 10 stations for tap out
top_tapout_stations = data_cleaned['tapOutStopsName'].value_counts().head(10)

plt.figure(figsize=(14,6))
top_tapin_stations.plot(kind='bar', color='skyblue', label='Tap In')
plt.title('10 Stasiun dengan Jumlah Tap In Terbanyak')
plt.xlabel('Nama Stasiun')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.legend()
plt.show()

plt.figure(figsize=(14,6))
top_tapout_stations.plot(kind='bar', color='salmon', label='Tap Out')
plt.title('10 Stasiun dengan Jumlah Tap Out Terbanyak')
plt.xlabel('Nama Stasiun')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# Analisis Berdasarkan Hari

```python
# Average travel duration based on day of the week
avg_duration_per_day = data_cleaned.groupby('day_name')['duration'].mean().reindex(days.values())

plt.figure(figsize=(14,6))
avg_duration_per_day.plot(kind='line', marker='o', color='green')
plt.title('Rata-Rata Durasi Perjalanan Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari')
plt.ylabel('Rata-Rata Durasi (menit)')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
